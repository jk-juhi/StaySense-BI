```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/archive/AB_NYC_2019.csv")

df.head()
```

```python
df.info()
df.shape
df.describe(include='all')
df.isnull().sum()
```

```python
#cleaning

# Create a copy for cleaning
airbnb_df = df.copy()

# Display first 5 rows
airbnb_df.head()
```

```python
#check missing values

airbnb_df.isnull().sum()
```

```python
#fill missing values

airbnb_df["name"] = airbnb_df["name"].fillna("Unknown Listing")

airbnb_df["host_name"] = airbnb_df["host_name"].fillna("Unknown Host")

airbnb_df["reviews_per_month"] = airbnb_df["reviews_per_month"].fillna(0)
```

```python
#convert data

airbnb_df["last_review"] = pd.to_datetime(
    airbnb_df["last_review"],
    errors="coerce"
)

airbnb_df.info()
```

```python
airbnb_df.duplicated().sum()
```

```python
#price category

def price_category(price):
    if price < 75:
        return "Budget"
    elif price < 200:
        return "Standard"
    elif price < 400:
        return "Premium"
    else:
        return "Luxury"

airbnb_df["price_category"] = airbnb_df["price"].apply(price_category)
```

```python
#review category

def review_category(reviews):
    if reviews == 0:
        return "No Reviews"
    elif reviews < 50:
        return "Low"
    elif reviews < 200:
        return "Medium"
    else:
        return "High"

airbnb_df["review_category"] = airbnb_df["number_of_reviews"].apply(review_category)
```

```python
#availability category

def availability(days):
    if days == 0:
        return "Unavailable"
    elif days <= 120:
        return "Low Availability"
    elif days <= 240:
        return "Medium Availability"
    else:
        return "High Availability"

airbnb_df["availability_category"] = airbnb_df["availability_365"].apply(availability)
```

```python
airbnb_df.head()
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)

airbnb_df.to_csv(
    "../data/processed/airbnb_cleaned.csv",
    index=False
)

print("Dataset saved successfully!")
```
