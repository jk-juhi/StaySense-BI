# Week 7 – Progress Report

## Objective
The objective of Week 7 was to develop prompt engineering skills for Business Intelligence by designing structured prompts, evaluating AI-generated responses, comparing different prompting techniques, and producing an AI-assisted Board-Level KPI Summary based on the StaySense BI project.

---

## Activities Completed

### 1. Created 10 Structured BI Prompts
Designed ten prompts covering different Business Intelligence tasks, including:
- Data cleaning
- KPI definition
- Executive summary generation
- SQL query generation
- Chart recommendation
- Correlation interpretation
- Data quality assessment
- Slide headline creation
- Business recommendation
- Explaining BI concepts to non-technical users

Each prompt was evaluated based on relevance, clarity, and usefulness.

---

### 2. Prompt Engineering Comparison
Compared three versions of the same executive summary prompt:
- Minimal Prompt
- Context-Rich Prompt
- Structured Role-Based Prompt

Observed how additional context, role assignment, and formatting constraints significantly improved the quality, accuracy, and professionalism of AI-generated responses.

---

### 3. AI-Generated KPI Summary
Generated a Board-Level KPI Summary using the StaySense BI dashboard insights and manually reviewed the output to ensure factual consistency with the NYC Airbnb 2019 dataset.

---

### 4. Fact Verification
Reviewed every generated statement against the project findings and corrected unsupported claims, resulting in a more accurate and business-focused executive report.

---

## Key Learning Outcomes

- Learned how prompt structure directly influences AI output quality.
- Understood the importance of providing context, audience, and constraints while prompting.
- Improved skills in evaluating AI-generated business content.
- Practiced fact-checking AI outputs using real project data.
- Gained experience in using Generative AI as a productivity tool for Business Intelligence reporting.

---

## Conclusion

Week 7 demonstrated that effective prompt engineering is an essential Business Intelligence skill. Carefully designed prompts consistently produced more accurate, relevant, and executive-ready outputs. Combining AI-generated content with human validation resulted in reliable business reports that support informed decision-making.
